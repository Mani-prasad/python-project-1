```python
# class defines a particular blueprint and an object is an instance/example of that particular class
# Attributes/properties
# Methods

# defining a class 'car'
class Car:
    # property 
    wheels = 4
    is_new = True
    headlights = 2
    speed = 30

    # syntax for constructing new object
    def __init__(self, model, color, mileage, fuel_type):
        self.model = model
        self.color = color
        self.mileage = mileage
        self.fuel_type = fuel_type

    def drive(self):
        print("Vroom vroom")
        self.is_new = False

    def description(self):
        print("It is a " + self.model + " of fuel type " + self.fuel_type + " and gives a mileage of " + self.mileage + " and the color is" + self.color)
    
    def acclerate(self):
        self.speed += 20
    

    def brake(self):
        self.speed -= 10

    def power_brake(self):
        self.speed -= 40
    
    def servicing(self):
        self.mileage = "30kmpl"


sonu_car = Car('audi','black',25,"petrol")
print(sonu_car.mileage)
sonu_car.servicing()
print(sonu_car.mileage)
```
