```python
#- Create a class, Rectangle. Its __init__() method should take self, angle1, angle2,
#angle3, angle4 as arguments
class Rectangle:
  number_of_sides=4
  angle_sum = 0
  def __init__(self,angle1,angle2,angle3,angle4):
    self.angle1=angle1
    self.angle2=angle2
    self.angle3=angle3
    self.angle4=angle4
  def check_angles(self,angle1,angle2,angle3,angle4):
    self.angle_sum+=(angle1+angle2+angle3+angle4)
    if self.angle_sum==360:
      print(True)
    else:
      print(False)

my_rectangle=Rectangle(100,30,40,50)
my_rectangle.check_angles(100,80,60,100)
#inheritence
class Square(Rectangle):
  angle=90
  def __init__(self,angle1,angle2,angle3,angle4):
    self.angle1=self.angle
    self.angle2=self.angle
    self.angle3=self.angle
    self.angle4=self.angle
my_square=Square(90,90,90,90)
print(my_square.angle)








     
      

      
   
```
