```python
class Vehicle:
  
  def __init__(self,name,max_speed,mileage):
    self.name=name
    self.max_speed=max_speed
    self.mileage=mileage
  def print_val(self):

    print('max_speed: ',self.max_speed)
    print('milega: ',self.mileage)
  def print_name_vehical(self):
    print('name: ',self.name)

```

```python
sonu=Vehicle('car',120,15)
sonu.print_val()
```

```python
class bus(Vehicle):
  def __init__(self,name,max_speed,mileage,model,color):

    self.model=model
    self.color=color

sonu_bus= bus("volvo",100,20,"modelx","Red")
print(sonu_bus.color)


```
