```python
# Create a 2d array with elements as 4,2,5 in 1st row and 3,7,5 in 2nd row
import numpy as np 
arr = np.array([[4,2,5],[3,7,5]])
print(arr)
```

```python
# Print type of arr object
print(type(arr))
```

```python
# Print array dimensions 
print(arr.ndim)
```

```python
# Print shape of array
print(arr.shape)
```

```python
# Print size (total number of elements) of array
print(arr.size)
```

```python
# Print type of elements in array
print(arr.dtype)
```

```python
# Create a sequence of 10 values in range 0 to 5
import numpy as np
a = np.arange(0,5,0.5)
a
```
