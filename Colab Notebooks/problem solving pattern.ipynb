```python
# armstrong number
for i in range(1001):
  num=i
  result=0
  n= len(str(i))
  while(i!= 0):
    digit=i%10
    result = result + digit**n
    i=i//10
  if num==result:
    print(num)

```

```python
# number pattern 
n=int(input("enter the no of rows:"))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end=" ")
  print()
```

```python
#printing star triangle 
n = int(input("enter the number:"))
for i in range(1,n+1):
  for j in range(1,i+1):
    print("*",end="")
  print()
```

```python
### hallow triangle 
n = int(input("enter the number:"))
for row in range(n):
  for col in range(n):
    if col==0 or row==(n-1) or row==col:
      print("*",end="")
    else:
      print(end=" ")
  print()
```
