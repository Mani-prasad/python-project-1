```python
def read_csv():
    with open('/content/births_usa.csv') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            print(row[0],row[1],row[4])
read_csv()
```

```python
def read_csv():
    with open('/content/births_usa.csv') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            print(row)
read_csv()
```

```python
def write_csv():
    with open('/content/sample.txt', 'w') as csvfile:
        writer = csv.writer(csvfile, delimiter=' ', quotechar='|', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(['id', 'name'])
        for i in range(10):
            writer.writerow([i, " january " + str(i),1500])
write_csv()
```

```python
# new method to write csv 
def write_csv():
    with open('/content/sample.txt', 'w') as csvfile1:
        fieldnames = ['id', 'name']
        writer = csv.DictWriter(csvfile1, fieldnames=fieldnames)
        writer.writeheader()
        for i in range(100):
            writer.writerow({'id': i, 'name': 'Masai ' + str(i)})
write_csv()
```
