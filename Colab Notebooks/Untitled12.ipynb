```python
#3. Write a program to use the loop to find the factorial of a given number.
#The factorial (symbol: !) means to multiply all whole numbers from the chosen number
# down to 1.
n=int(input())
factorial=1
for i in range(1,n+1):
  factorial=factorial*i
print(factorial)

  



```

```python
#6. Write a program to calculate the sum of series up to n term. For example, if n =5 the series
#will become 2 + 22 + 222 + 2222 + 22222 = 24690.
num = input()
ran=int(input())
sum1=0
for i in range(1,ran+1):
  temp=num*i
  sum1=sum1+int(temp)
print(sum1)

```
