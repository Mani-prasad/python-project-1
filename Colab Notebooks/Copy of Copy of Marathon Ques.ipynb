10. .Create a child class Bus that will inherit all of the variables and methods of the Vehicle class.
also print the speed and mileage of the bus.


```python
class bus(Vehicle):
  def __init__(self,max_speed,mileage,model):
    super().__init__('bus',max_speed,mileage)
    self.model=model
  def print_val(self):
    super().print_val()
    print('model: ',self.model)

    
```

```python
c=bus(120,5,'mini') #bus,120,5,mini
```

```python
c.print_val()
```
