```python
with open("/content/births_usa.csv", 'r') as f:
  reader = csv.reader(f)
  # read column names
  cols = next(iter(reader))
  # read rows
  rows = []
  for i in reader:
    rows.append(i)
```

```python
months_list = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
          
```

```python
d = dict.fromkeys(months_list, 0)
d
```

```python
for row in rows:
  month_idx = int(row[1])
  birth_count = int(row[-1])

  # update in dictionary
  # month starts with 1 and ends with 12
  # but index starts with 0 and ends with 11
  month_name = months_list[month_idx-1]
  d[month_name] += birth_count 
```

```python
for month in months_list:
  print(month, d[month], sep=" - ")
```

```python
# write report in a CSV file
write_path = "/content/births_usa.csv"

with open(write_path,'w') as f:
  writer = csv.writer(f)
  # write column names
  writer.writerow(['month', 'total_births'])
  # write data
  for month in months_list:
    writer.writerow([month, d[month]])
```

```python
path = "/content/temperatures.csv"

with open(path, 'r') as f:
  reader = csv.reader(f)
  cols = next(iter(reader))
  rows = []

  for i in reader:
    rows.append(i)

  print(len(rows), len(cols))
```

```python
# what columns are there?
cols
```

```python
# how a row looks like?
rows[0]
```

```python
# we are interested in month, year, actual_min_temp, actual_max_temp
summary = {}

for row in rows:
  date = row[0]
  year,month,day = map(int,date.split("-"))

  actual_min_temp = int(row[2])
  actual_max_temp = int(row[3])

  # add necessary data
  # if year, month pair is new, create a key-value pair
  if (year, month) not in summary:
    summary[(year, month)] = [actual_min_temp, actual_max_temp]    

  else:
    # compare and update
    if actual_min_temp < summary[(year, month)][0]:
      summary[(year, month)][0] = actual_min_temp 

    if actual_max_temp > summary[(year, month)][1]:
      summary[(year, month)][1] = actual_max_temp
    
len(summary)
```

```python
# write in a CSV file
write_path = "/content/temperatures.csv"
with open(write_path, 'w') as f:
  writer = csv.writer(f)

  # write column headers
  new_cols = ["year", "month", "actual_min_temp", "actual_max_temp"]
  writer.writerow(new_cols)
  print(new_cols)
  # write data
  for key,val in sorted(summary.items()): 
    data = [key[0], key[1], val[0], val[1]]
    writer.writerow(data)
    print(data)
```
