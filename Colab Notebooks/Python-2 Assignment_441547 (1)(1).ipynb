Given the list, reverse the list but with out the last elements using slicing.

- input : l = [3,4,8,9,4,3]

- output: [4,9,8,4,3]

```python
l = [3,4,8,9,4,3]
l[ -2: :-1]
```

Using slicing, achieve the following output as mentioned below.

- input: l = ['Pushap', 'Anna', 7, 8, 9, [0,1]]

- output: 'annA'

```python
l = ['Pushap', 'Anna', 7, 8, 9, [0,1]]
l[1][::-1]
```
