```python
#Calculate square loss and Absolute Error lo
Actual = [2,6,1,7,-9,5]


Real = [5,8,3,12,-9,0]
n = 6
sum = 0
  
# for loop for iteration
for i in range(n):
    sum += abs(Actual[i] - Real[i])
  
error = sum/n
  
# display
print("Mean absolute error : " + str(error))
```
