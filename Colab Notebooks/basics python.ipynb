```python
# armstrong number 
n = int(input())
sum1=0
a = len(str(n))
while (n>0):
  sum1 = sum1 + (n%10)**a
  if sum1 == n :
    print("yes armstrong number")
  n = n//10
else:
  print("no")

 
```

```python
n = int(input())
i = n 
count = 0 
while (i>0):
  count= count+1
  i=i//10
sum1 = 0 
i = n 
while (i>0):
  digit=i%10
  x = 1 
  pro = 1
  while(x<=count):
    pro = pro*digit
    x = x+1
  sum1 = sum1 + pro
  i = i//10
if (sum1==n):
  print("armstrong")
else:
  print("non armstrong")
```
