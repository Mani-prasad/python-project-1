```python
###1. Write a NumPy program to compute the multiplication of two given matrices
a = np.arange(1,6)
b = np.arange(6,11)
print(a*b)

```

```python
##2. Write a NumPy program to compute the outer product of two given vectors.
a = np.arange(1,6)
b = np.arange(6,11)
result = np.outer(a,b)
print(result)
print(a)
print(b)
```

```python
#Q3 3. Write a NumPy program to compute the cross product of two given vectors.
c = np.array([1,2,3])
d = np.array([4,5,6])
result = np.cross(c,d)
print(result)
```

```python
###4. Write a NumPy program to compute the determinant of a given square array.
a = np.array([[1, 0], [1, 2]])
print(a)
print(np.linalg.det(a))
```

```python
#5. Write a NumPy program to evaluate Einstein's summation convention of two given
#multidimensional arrays.
#
a = np.array([1,2,3])
b = np.array([4,5,6])
result =  np.einsum("n,n", a, b)
print(result)


```

```python
#6. Write a NumPy program to compute the inner product of vectors for 1-D arrays (without
#complex conjugation)
#and in higher dimension.
a = np.array([1,2,3])
b = np.array([4,5,6])
result = np.inner(a, b)
print(result)
```

```python
#7. Write a NumPy program to compute the eigenvalues and right eigenvectors of a given
#square array.
m = np.mat("4 -2;1 2")
w, v = np.linalg.eig(m) 
print( "Eigenvalues of the said matrix",w)
print( "Eigenvectors of the said matrix",v)
```

```python
#8. Write a NumPy program to compute the Kronecker product of two given mulitdimension
#arrays.
a = np.array([1,2,3])
b = np.array([4,5,6])

result =  np.kron(a, b)
print(result)
```

```python
###9. Write a NumPy program to compute the condition number of a given matrix.
import numpy as np
from numpy import linalg as LA
a = np.array([[1, 2, -3], [4, 1, 5], [1, 0, 1]])
print("Original matrix:")
print(a.shape)
print(a.ndim)
print("The condition number of the said matrix:")
print(LA.cond(a))
```

```python
#10. Write a NumPy program to find a matrix or vector norm.
v = np.arange(7)
result = np.linalg.norm(v)
print("Vector norm:")
print(result)
m = np.matrix('1, 2; 3, 4') 
result1 = np.linalg.norm(m)
print("Matrix norm:")
print(result1)
```

```python
#12. Write a NumPy program to compute the inverse of a given matrix
m = np.array([[1,2],[3,4]])

print(m)
result =  np.linalg.inv(m)

print(result)
```

```python
#13. Write a NumPy program to calculate the QR decomposition of a given matrix.
m = np.array([[1,2],[3,4]])
print("Original matrix:")
print(m)
print(m.ndim)
result =  np.linalg.qr(m)
print("Decomposition of the said matrix:")
print(result)
```
